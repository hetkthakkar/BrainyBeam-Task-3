### Normalization

```python
# Select numerical columns for normalization
numerical_cols = ['Unit_Cost', 'Total_Revenue', 'Total_Profit']
numerical_data = data[numerical_cols]

# Min-Max Scaling
min_max_scaler = MinMaxScaler()
min_max_scaled = min_max_scaler.fit_transform(numerical_data)

# Z-Score Normalization
z_score_scaler = StandardScaler()
z_score_scaled = z_score_scaler.fit_transform(numerical_data)

# Decimal Scaling
decimal_scaled = numerical_data / (10 ** np.ceil(np.log10(numerical_data.abs().max())))

# Log Transformation
log_transformed = np.log1p(numerical_data)

# Combine the normalized results into a new DataFrame
normalized_data = pd.DataFrame({
    'Unit_Cost_MinMax': min_max_scaled[:, 0],
    'Total_Revenue_MinMax': min_max_scaled[:, 1],
    'Total_Profit_MinMax': min_max_scaled[:, 2],
    'Unit_Cost_ZScore': z_score_scaled[:, 0],
    'Total_Revenue_ZScore': z_score_scaled[:, 1],
    'Total_Profit_ZScore': z_score_scaled[:, 2],
    'Unit_Cost_Decimal': decimal_scaled['Unit_Cost'],
    'Total_Revenue_Decimal': decimal_scaled['Total_Revenue'],
    'Total_Profit_Decimal': decimal_scaled['Total_Profit'],
    'Unit_Cost_Log': log_transformed['Unit_Cost'],
    'Total_Revenue_Log': log_transformed['Total_Revenue'],
    'Total_Profit_Log': log_transformed['Total_Profit']
})

# Display the first few rows of the normalized data
print(normalized_data)
```

### Z-Score

```python
# Select numerical columns for Z-Score normalization and mean centering
numerical_cols = ['Unit_Cost', 'Total_Revenue', 'Total_Profit']
numerical_data = data[numerical_cols]

# Z-Score Normalization (Mean-Centering included automatically)
scaler = StandardScaler()
z_score_scaled = scaler.fit_transform(numerical_data)

# Convert the result back to a DataFrame with mean-centered values
z_score_normalized_data = pd.DataFrame(z_score_scaled, columns=numerical_cols)

# Display the first few rows of the Z-Score normalized (mean-centered) dataset
print(z_score_normalized_data)
```
